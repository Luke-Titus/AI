```python
from sklearn.datasets import load_breast_cancer
from sklearn.datasets import make_blobs
from sklearn.linear_model import Lasso
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import Ridge
from sklearn.model_selection import train_test_split
from sklearn.svm import LinearSVC

import mglearn
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Generate some data and an example fit to the data
mglearn.plots.plot_linear_regression_wave()
```

```python
# Create this model by training a LinearRegression model using ordinary least squares
X,y=mglearn.datasets.make_wave(n_samples=60)
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=42)
model=LinearRegression()
lr=model.fit(X_train,y_train)
print("lr.coef_:",lr.coef_)
print("lr.intercept_:",lr.intercept_)
```

```python
# Plot the fitted model with the data
plt.scatter(X,y,c="red")
line=np.linspace(-3,3,1000).reshape(-1,1)
plt.plot(line,model.predict(line))
plt.show()
```

```python
# Fit a linear model to the Boston Housing dataset, which has 506 samples and 104 features each sample. 
X, y = mglearn.datasets.load_extended_boston()
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=0)
model=LinearRegression()
lr=model.fit(X_train,y_train)
print("Training set score: {:.2f}".format(lr.score(X_train,y_train)))
print("Test set score: {:.2f}".format(lr.score(X_test,y_test)))
```

```python
# Ridge regression
# Fit a linear model with regularization to avoid overfitting.
# By default, Ridge regression uses L2 regularization, which penalizes the square of the coefficients. 
model=Ridge()
ridge=model.fit(X_train, y_train)
print("Training set score: {:.2f}".format(ridge.score(X_train,y_train)))
print("Test set score: {:.2f}".format(ridge.score(X_test,y_test)))

# Note that the training set score is lower but the test score is higher. This is an indication that the model did not overfit the data. 
```

```python
# Use the alpha parameter to tune how much regularization is used with Ridge regression
model=Ridge(alpha=10)
ridge10=model.fit(X_train,y_train)
print("Training set score: {:.2f}".format(ridge10.score(X_train,y_train)))
print("Test set score: {:.2f}".format(ridge10.score(X_test,y_test)))
```

```python
model=Ridge(alpha=0.1)
ridge01=model.fit(X_train,y_train)
print("Training set score: {:.2f}".format(ridge01.score(X_train,y_train)))
print("Test set score: {:.2f}".format(ridge01.score(X_test,y_test)))
```

```python
# Visualize the magnitude of the model coefficients for each value of alpha
# For a higher value of alpha, the magnitude of the parameters is lower, which means lower model complexity, so better generalization.
plt.plot(ridge.coef_,'s',label="Ridge alpha=1")
plt.plot(ridge10.coef_,'^',label="Ridge alpha=10")
plt.plot(ridge01.coef_,'v',label="Ridge alpha=0.1")
plt.plot(lr.coef_,'o',label="LinearRegression")
plt.xlabel("Coefficient index")
plt.ylabel("Coefficient magnitude")
plt.hlines(0,0,len(lr.coef_))
plt.ylim(-25,25)
plt.legend()
plt.show()
```

```python
# Lasso Regression
# Lasso regression is like Ridge regression, except the L1 norm is used.
# Rather than L2 regularization which tries to make the coefficients as small as possible, 
#   L1 allows for some coefficients to be exactly zero.

model=Lasso()
lasso=model.fit(X_train,y_train)
print("Traing set score: {:.2f}".format(lasso.score(X_train,y_train)))
print("Test set score: {:.2f}".format(lasso.score(X_test,y_test)))
print("Number of features used:",np.sum(lasso.coef_!=0))

# Note that the training and test scores are bad, meaning that the model is underfitting the data.
```

```python
# Use the alpha parameter to tune how much regularization is performed. 
model=Lasso(alpha=0.01,max_iter=100000).fit(X_train,y_train)
lasso001=model.fit(X_train,y_train)
print("Traing set score: {:.2f}".format(lasso001.score(X_train,y_train)))
print("Test set score: {:.2f}".format(lasso001.score(X_test,y_test)))
print("Number of features used:",np.sum(lasso001.coef_!=0))
```

```python
# Note that the smaller alpha becomes, the fewer coefficients are zero. 
# Thus, a lower value of alpha allows for a more complex model to be used.
model=Lasso(alpha=0.0001,max_iter=100000).fit(X_train,y_train)
lasso00001=model.fit(X_train,y_train)
print("Traing set score: {:.2f}".format(lasso00001.score(X_train,y_train)))
print("Test set score: {:.2f}".format(lasso00001.score(X_test,y_test)))
print("Number of features used:",np.sum(lasso00001.coef_!=0))
```

```python
# Visualize the magnitude of the model coefficients for each value of alpha
plt.plot(lasso.coef_,'s',label="Lasso alpha=1")
plt.plot(lasso001.coef_,'^',label="Lasso alpha=0.01")
plt.plot(lasso00001.coef_,'v',label="Lasso alpha=0.0001")

plt.plot(ridge01.coef_,'o',label="Ridge alpha=0.1")
plt.legend(ncol=2,loc=(0,1.05))
plt.ylim(-25,25)
plt.xlabel("Coefficient index")
plt.ylabel("Coefficient magnitude")
plt.show()
```

```python
# Logistic Regression & Support Vector Machines for classification
# Linear models for classification try to create a decision boundary between the various classes.
X, y = mglearn.datasets.make_forge()
fig, axes=plt.subplots(1,2,figsize=(10,3))
for model, ax in zip([LinearSVC(max_iter=100000, dual=True),LogisticRegression(max_iter=100000)],axes):
	clf=model.fit(X,y)
	mglearn.plots.plot_2d_separator(clf,X,fill=False,eps=0.5,ax=ax,alpha=0.7)
	mglearn.discrete_scatter(X[:,0],X[:,1],y,ax=ax)
	ax.set_title(clf.__class__.__name__)
	ax.set_xlabel("Feature 0")
	ax.set_ylabel("Feature 1")
axes[0].legend()
plt.show()
```

```python
# The parameter C controls regularization
# A higher value of C corresponds to less regularization.
# For a high value of C, linear models try to fit the training set
# For a low value of C, the model puts more emphasis on finding coefficients that are close to zero. 
mglearn.plots.plot_linear_svc_regularization()
```

```python
# Analyze Logistic Regression using the Breast Cancer dataset which has more features per sample. 
cancer=load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(cancer.data,cancer.target,stratify=cancer.target,random_state=42)
logreg=LogisticRegression(max_iter=100000).fit(X_train,y_train)
print("Training set score: {:.3f}".format(logreg.score(X_train,y_train)))
print("Test set score: {:.3f}".format(logreg.score(X_test,y_test)))

# Since the training and test set scores are quite close, we are likely underfitting. 
```

```python
# To prevent underfitting, increase C to reduce regularization. 
logreg100=LogisticRegression(max_iter=100000,C=100).fit(X_train,y_train)
print("Training set score: {:.3f}".format(logreg100.score(X_train,y_train)))
print("Test set score: {:.3f}".format(logreg100.score(X_test,y_test)))
```

```python
# Compare with reducing C to increase regularization. 
logreg001=LogisticRegression(max_iter=100000,C=0.01).fit(X_train,y_train)
print("Training set score: {:.3f}".format(logreg001.score(X_train,y_train)))
print("Test set score: {:.3f}".format(logreg001.score(X_test,y_test)))
```

```python
# Visualize the magnitude of the coefficients for each model.
plt.figure(figsize=(4,3))
plt.plot(logreg.coef_.T,'o',label="C=1")
plt.plot(logreg100.coef_.T,'^',label="C=100")
plt.plot(logreg001.coef_.T,'v',label="C=0.01")
plt.xticks(range(cancer.data.shape[1]),cancer.feature_names,rotation=90)
plt.hlines(0,0,cancer.data.shape[1])
plt.ylim(-5,5)
plt.xlabel("Feature")
plt.ylabel("Coefficient magnitude")
plt.legend()
plt.show()
```

```python
# Multiclass classification
# A common technique with linear models is to use a one-vs-rest approach. This uses a binary model is learned
#   for each class that tries to separate that class from the others. 

# Consider a different dataset with three classes. 
X,y=make_blobs(random_state=42)
mglearn.discrete_scatter(X[:,0],X[:,1],y)
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
plt.legend(["Class 0", "Class 1", "Class 2"])
plt.show()
```

```python
# Fit a linear support vector machine model to the blob data.
linear_svm=LinearSVC(dual=True).fit(X,y)
print("Coefficient shape: ",linear_svm.coef_.shape)
print("Intercept shape: ", linear_svm.intercept_.shape)
```

```python
# Visualize the decision boundaries for each class. 
mglearn.discrete_scatter(X[:,0],X[:,1],y)
line=np.linspace(-15,15)
for coef, intercept, color in zip(linear_svm.coef_,linear_svm.intercept_,mglearn.cm3.colors):
	plt.plot(line,-(line*coef[0]+intercept)/coef[1],c=color)
plt.ylim(-10,15)
plt.xlim(-10,8)
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
plt.legend(['Class 0','Class 1', 'Class 2', 'Line class 0','Line class 1','Line class 2'],loc=(1.01,0.3))
plt.show()
```

```python
# For any point that lies in the triangle in the middle, the class will be assigned based on the model with the 
#   highest value for the classification formula w[0]*x[0]+w[1]*x[1]+...+b. This corresponds to the class of the closest line.
```
